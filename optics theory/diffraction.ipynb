
## Helmholtz–Kirchhoff diffraction integral

Using Greens 2. integral formula with the wave amplitude $\Phi=U$ and the green function (helper function) $\Psi = G$ we get:


\begin{align}
\int_V \big( U \vec{\nabla}^2 G  - G \vec{\nabla}^2 U \big) \; d^3V &= \oint_{S} \Big( U \frac{\partial G}{\partial n} - G \frac{\partial U}{\partial n} \Big) \; d^2S \\
\end{align}

since $\vec{\nabla}^2 G$ and $\vec{\nabla}^2 U$ are equal to $-k^2$ the left side can be rewritten as $-k^2 \int_V \big( U G  - GU \big) \; d^3V = 0$, wich is equal to zero.

The result is, that the right side is an integration over an arbitrary closed Surface, except point $\vec{r}=\vec{0}$ since the functions are discontinous there. We split the right integral into the integration over a surface S minus (inverse surface normal) the integration over a sphere D of radius $\epsilon$ around point $\vec{0}$. To summurazie so far we get:

\begin{align}
&0 = \oint_{S} \Big( U \frac{\partial G}{\partial n} - G \frac{\partial U}{\partial n} \Big) \; d^2S - \oint_{D} \Big( U \frac{\partial G}{\partial n} - G \frac{\partial U}{\partial n} U \Big) \; d^2S \\
\Leftrightarrow & \oint_{D} \Big( U \frac{\partial G}{\partial n} - G \frac{\partial U}{\partial n} U \Big) \; d^2S = \oint_{S} \Big( U \frac{\partial G}{\partial n} - G \frac{\partial U}{\partial n} \Big) \; d^2S \\
\end{align}


Then we replace $G = \frac{e^{ik|\vec{r}|}}{|\vec{r}|}$. On the left side we apply $ \frac{\partial G}{\partial n} = \vec{n} \cdot \vec{\nabla} G = \vec{n} \cdot \vec{e}_r \frac{\partial}{\partial \vec{r}} \frac{e^{ik|\vec{r}|}}{|\vec{r}|} = \big( ik - \frac{1}{|r|} \big) \frac{e^{ik|\vec{r}|}}{|\vec{r}|}$, since $\vec{n}$ is parallel to $\vec{e}_r$. Than we rewrite the integration on the left side in spherical coordinates $\int ..... d^2S = \int_0^{4\pi} ..... \vec{r}^2 d\Omega$ and replace $|\vec{r}|=\epsilon$:

\begin{align}
\int_0^{4\pi} \bigg( U\big( ik - \frac{1}{\epsilon} \big) \frac{e^{ik\epsilon}}{\epsilon} - \frac{e^{ik\epsilon}}{\epsilon} \frac{\partial U}{\partial n} \bigg) \epsilon^2 d\Omega &= \int_S \bigg( U \frac{\partial}{\partial n} \Big( \frac{e^{ik|\vec{r}|}}{|\vec{r}|} \Big) - \frac{e^{ik|\vec{r}|}}{|\vec{r}|} \frac{\partial U}{\partial n} \bigg) d^2S
\end{align}

The left side is equal to $\int_0^{4\pi} e^{ik\epsilon} \bigg( U\big( ik \epsilon - 1 \big) - \epsilon \frac{\partial U}{\partial n} \bigg) d\Omega$. Taking the limit $\epsilon \rightarrow 0$ we get $\int_0^{4\pi} U(\vec{0}) d\Omega = 4\pi U(\vec{0})$. From this follows:

\begin{align}
U(\vec{0}) &= \frac{1}{4\pi} \int_S \bigg( U \frac{\partial}{\partial n} \Big( \frac{e^{ik|\vec{r}|}}{|\vec{r}|} \Big) - \frac{e^{ik|\vec{r}|}}{|\vec{r}|} \frac{\partial U}{\partial n} \bigg) d^2S
\end{align}

<img src="kirchhoff.png" width="400px">

Shifting our point $\vec{0}$ to an arbitrary Point P produces the Kirchhoff diffraction formula (shifting is done after derivation for simplicity of the derivation. The green function is positioned independent of the geometry of the problem and for that reason could've been shifted before derivation wich would produce the same result). The length of vector $\vec{r}$ is replaced by distance $s$ between P and points on the surface of integration (aperture). This is the Helmholtz–Kirchhoff diffraction integral.

<font color='red'>

\begin{align}
U(P) &= \frac{1}{4\pi} \int_S \bigg( U \frac{\partial G}{\partial n} - G \frac{\partial U}{\partial n} \bigg) d^2S \\
&= \frac{1}{4\pi} \int_S \bigg( U \frac{\partial}{\partial n} \Big( \frac{e^{iks}}{s} \Big) - \frac{e^{iks}}{s} \frac{\partial U}{\partial n} \bigg) d^2S
\end{align}

</font>

It states, that the field (disturbance) at Position P is completly determined by the amplitude $U$ and change $\frac{\partial U}{\partial n}$ through a closed surface around that point.



### Fresnel–Kirchhoff diffraction formula

When integrating around a point P behind an aperture, we can split the surface integral into three surfaces A,B,C. Light amplitude at B is blocked by the aperture and at C also zero, if the light is turned on at time $t_0$ and radius $R > c \cdot t$ chosen to be outside of the field. Discontinuties at the boundries are neglegted. The result is an integral only over the aperture opening A with the distance s prom point P to points Q at the aperture.

\begin{align}
U(P) &= \frac{1}{4\pi} \int_A \bigg( U \frac{\partial}{\partial n} \Big( \frac{e^{iks}}{s} \Big) - \frac{e^{iks}}{s} \frac{\partial U}{\partial n} \bigg) d^2S
\end{align}

Now we consider the special case of a point source at point $P_0$ wich shines light to points Q at a distance r. Then the light is transmitted to point P at the distance s. The intensity drops additionally with the cosine to the surface normal $\vec{n}$.

\begin{align}
G = \frac{e^{iks}}{s} \qquad \frac{\partial G}{\partial n} = (iks - 1)\frac{e^{iks}}{s} \cos(\vec{n},\vec{s}) \\
U = \frac{e^{ikr}}{r} \qquad \frac{\partial U}{\partial n} = (ikr - 1)\frac{e^{ikr}}{r} \cos(\vec{n},\vec{r})
\end{align}

\begin{align}
U(P) &= \frac{1}{4\pi} \int_A \frac{e^{ik(r+s)}}{rs} \bigg( (iks - 1)\cos(\vec{n},\vec{s}) - (ikr - 1) \cos(\vec{n},\vec{r}) \bigg) d^2S
\end{align}

At greater distance from the aperture $k>>1/s$ and $k>>1/r$ the equation simplifies by means of $(iks-1)/s = (ik-1/s) \approx ik$ and $(ikr-1)/r \approx ik$. Also $k=2\pi/\lambda$ was used.

<font color='red'>

\begin{align}
U(P) &= \frac{1}{ik} \int_A \frac{e^{ik(r+s)}}{r+s} \bigg( \frac{\cos(\vec{n},\vec{r}) - \cos(\vec{n},\vec{s})}{2} \bigg) d^2S
\end{align}

</font>


## Huygens–Fresnel equation (spherical waves)

If we consider $\cos(\vec{n},\vec{r}) \approx 1$ for small apertures and a distant point source $P_0$ and using $(\vec{n},\vec{s}) = \pi - \gamma$, we get the Huygens–Fresnel equation.

\begin{align}
U(P) &= \frac{1}{ik} \int_A \frac{e^{ik(r+s)}}{r+s} \bigg( \frac{1 + \cos(\gamma)}{2} \bigg) d^2S \\
&= \frac{e^{ikr}}{ikr} \int_A \frac{e^{iks}}{iks} \bigg( \frac{1 + \cos(\gamma)}{2} \bigg) d^2S
\end{align}

This equation discribes the light amplitude at point P as emerging sperical waves at the aperture weighted with $\frac{1 + \cos(\gamma)}{2} \in [0,1]$ to account for the intensity drop for emitting light at an angle. They propagate with $\frac{e^{ikr}}{ikr}$ through free space.


## Rayleigh–Sommerfeld diffraction equation

The Fresnel–Kirchhoff diffraction formula assumes, that at surface B: $U=\frac{\partial U}{\partial n}=0$, from wich follows $U=constant$ and since $U$ is continuous on the boundaries, $U=0$ everywhere follows. This is obviously a contradiction. An alternative is to assume $U=0$ or $\frac{\partial U}{\partial n}=0$ over the entire screen (A+B). With the Helmholtz–Kirchhoff diffraction integral

\begin{align}
U(P) &= \frac{1}{4\pi} \int_S \bigg( U \frac{\partial G}{\partial n} - G \frac{\partial U}{\partial n} \bigg) d^2S
\end{align}

and a modified helper function 

\begin{align}
G = \frac{e^{iks}}{s} - \frac{e^{ik\overline{s}}}{\overline{s}}
\end{align}

where $\overline{s}$ is the vector $s$ mirrored at the plane of the aperture. We can derive a mathematically more rigorous diffraction equation.
From our new helper function $G$ follows:

\begin{align}
G \Big|_{A,B} &= 0 \\
\frac{\partial G}{\partial n} \Big|_{A,B} &= 2 \bigg( -ik-\frac{1}{s} \bigg) \frac{e^{-iks}}{s} \cos(\vec{n},\vec{s})
\end{align}

inserting this into the Helmholtz–Kirchhoff diffraction integral (with $k>>1/s$ and $\lambda=2\pi/k$) results in the first Rayleigh–Sommerfeld diffraction equation

<font color='red'>

\begin{align}
U(P) &= \frac{1}{i\lambda} \int_{A,B} U \frac{e^{-iks}}{s} \cos(\vec{n},\vec{s}) d^2S
\end{align}

</font>

Another solution is to use following helper function

\begin{align}
G = \frac{e^{iks}}{s} + \frac{e^{ik\overline{s}}}{\overline{s}}
\end{align}

From this one follows:

\begin{align}
G \Big|_{A,B} &= 2 \frac{e^{-iks}}{s} \\
\frac{\partial G}{\partial n} \Big|_{A,B} &= 0
\end{align}

Wich inserted into the Helmholtz–Kirchhoff diffraction integral results in

<font color='red'>

\begin{align}
U(P) &= \frac{1}{2\pi} \int_S \frac{e^{-iks}}{s} \frac{\partial U}{\partial n} d^2S
\end{align}

</font>

## Fresnel diffraction equation

starting with the first Rayleigh–Sommerfeld diffraction equation, we can simplify $\cos(\vec{n},\vec{s}) \approx z/s$, where z is the distance from a point source to the screen (along optical axis z).

\begin{align}
U(P) &= \frac{z}{i\lambda} \int_{A,B} U \frac{e^{-iks}}{s^2} d^2S
\end{align}

Then the radius vector s is approximated as:

\begin{align}
s = \sqrt{(x-u)^2 + (y-v)^2 + z^2} &= z \sqrt{\bigg(\frac{x-u}{z}\bigg)^2 + \bigg(\frac{y-v}{z}\bigg)^2 + 1} \\
&\approx z \bigg( \frac{1}{2} \bigg(\frac{x-u}{z}\bigg)^2 + \frac{1}{2} \bigg(\frac{y-v}{z}\bigg)^2 + 1 \bigg) \\
&= z \bigg( \frac{1}{2} \bigg(\frac{x-u}{z}\bigg)^2 + \frac{1}{2} \bigg(\frac{y-v}{z}\bigg)^2 + 1 \bigg) \\
&= z + \frac{1}{2z} \Big( (x-u)^2 + (y-v)^2 \Big)
\end{align}

where u and v are positions of the point source along the x and y axis. 
For the radius vector in the denominator we can even use $s=z$, because a small change in space is not as crucial as a small change in the phase $e^{-iks}$. The result is the Fresnel diffraction equation:

<font color='red'>

\begin{align}
U(P) &= \frac{z}{i\lambda} \int_{A,B} U \frac{e^{-ik \bigg( z + \frac{1}{2z} \Big( (x-u)^2 + (y-v)^2 \Big) \bigg)}}{z^2} d^2S \\
&= \frac{e^{-ikz}}{i z \lambda} \int_{A,B} U e^{-\frac{ik}{2z} \big( (x-u)^2 + (y-v)^2 \big)} d^2S \\
&= \frac{e^{-ikz}}{i z \lambda} e^{-\frac{ik}{2z} \big( x^2 + y^2 \big)} \int_{A,B} \Bigg[ U(u,v) e^{-\frac{ik}{2z} \big( u^2 + v^2 \big)} \Bigg] e^{+\frac{ik}{2z} \big( xu + yv \big)} dudv
\end{align}

</font>

The last version of this equation makes emphasize on the fact, that the Fresnel diffraction is the fourier transform of the light intensity U multiplied with a quadratic phase factor $e^{-\frac{ik}{2z} \big( u^2 + v^2 \big)}$.


## Fraunhofer diffraction (Fourier Transform)

For very far distances $z>>\frac{k}{2}(u^2+v^2)$ this simplifies further to the <b>fourier transform</b>

<font color='red'>

\begin{align}
U(P) &= \frac{e^{-ikz}}{i z \lambda} e^{-\frac{ik}{2z} \big( x^2 + y^2 \big)} \int_{A,B} U(u,v) e^{+\frac{ik}{2z} \big( xu + yv \big)} dudv
\end{align}

</font>
