## Propagation of Wigner function


#### propagation trough free space

\begin{align}
H(k_x,k_y) &= e^{i k_z z} = e^{i \sqrt{k^2-k_x^2-k_y^2} \cdot z} \approx e^{ikz \big( 1-\frac{k_x^2 + k_y^2}{2k^2} \big)} \approx e^{\frac{-iz}{2k} \big( k_x^2 + k_y^2 \big)} \\
\end{align}

wich resuts in a shear of the wigner distribution alogn the $x$-axis when signal $U(x,y)$ is modified by $H(x,y)$:

\begin{align}
K(x_o,k_o,x_i,k_i) = \delta(k_i-k_o) \delta( x_i - x_o - k_o \frac{z}{k})
\end{align}

\begin{align}
W(x,k_x) \rightarrow W \big( x - k_x \frac{z}{k} , k_x \big)
\end{align}

#### propagation through lense (circular phase change):

\begin{align}
h(x,y) = e^{-ik \frac{x^2+y^2}{2 f}}
\end{align}

wich resuts in a shear of the wigner distribution along the $q$-axis when signal $U(x,y)$ is modified by $H(x,y)$:

\begin{align}
K(x_o,k_o,x_i,k_i) = \delta(x_i-x_o) \delta( k_i - k_o - x_o \frac{k}{f} )
\end{align}

\begin{align}
W(x,k_x) \rightarrow W \big( x , k_x + x \frac{k}{f} \big)
\end{align}


## First order optical system

propagation functions $h(x)$ have constant absolute value and phases vary quadratically.

\begin{align}
K(x_o,k_o,x_i,k_i) = 2 \pi \gamma \delta(x_i - A x_o - B k_o) \delta( k_i - C x_o - D k_o)
\end{align}

with $\gamma = 1$ for lossless systems. This result in a wigner function

\begin{align}
W_o(x,k) = W_i(Ax+Bk,Cx+Dk)
\end{align}

This discription is equivalent to the ray discription with ABCD matricies:

\begin{align}
\begin{pmatrix} x_i \\ k_i \end{pmatrix} = \begin{pmatrix} A & B \\ C & D \end{pmatrix} \begin{pmatrix} x_o \\ k_o \end{pmatrix}
\end{align}



## fractional fourier transform

We consider the Wigner domain (x,k), where the projection of the wigner function onto the space axis $x$:

\begin{align}
|I(r)|^2 = \int W(r,q) dq
\end{align}

results in the intensity and the projection onto the frequency axis $k$:

\begin{align}
|I(q)|^2 = \int W(r,q) dr
\end{align}

results in the frequency spectrum, wich is the fourier transform of the former.
Hence a rotation by 90Â° in wigner space is the normal fourier transform.

\begin{align}
\mathcal{F}\big\{ W(r,q) \big\} = \mathcal{F}\big\{ W \big\} (-q,r)
\end{align}

The fractional fourier transform can be derived from this principle as a rotation by a fraction of this angle.

## optical implementation

#### wigner space rotation

by 3 consecutive shear operations:<br>
free space -> lens -> free space;  with $R=\frac{z}{k}=\tan(\frac{\phi}{2})$ and $Q=\frac{k}{f}=\sin(\phi)$ <br>
lens -> free space -> lens;  with $R=\frac{z}{k}=\sin(\phi)$ and $Q=\frac{k}{f}=\tan(\frac{\phi}{2})$ <br>

where the fractional degree $P$ determines the rotation angle $\phi = P \pi / 2$

#### combining these equations

\begin{align}
\mathcal{F}_P \Big\{ u(x) \Big\} &= \iint_{-\infty}^{\infty} u_0(x') \underbrace{e^{-ik \frac{x^2}{2 f}}}_{lens} e^{-ik_x x'} dx' \underbrace{e^{\frac{-iz}{2k} k_x^2}}_{free space} e^{i k_x x} dk_x \underbrace{e^{-ik \frac{x^2}{2 f}}}_{lens} \\
&= ... \\
&= \int_{-\infty}^{\infty} u(x') e^{i \frac{k}{2f} \tan(\phi) \big( x^2 + x'^2 \big)} e^{-i \frac{k}{2f} \sin(\phi) x x' } dx'
\end{align}
