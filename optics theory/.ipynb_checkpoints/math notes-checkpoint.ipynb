## Convolution
multiply function $f(t)$ with impulse response $g(\tau-t)$ for all times t. Impulse response $g(\tau-t)$ is flipped auround y axis (negative sign before t), when comparing to $g(t)$. This operation is motivated by a simple multiplication with the impulse response $g(\omega)$ in the fourier domain.<br>
application: convolving signal with impulse response

\begin{align}
(f \circledast g)(\tau) &= \int_{-\infty}^{-\infty} f(t)g(\tau-t) dt \\
&= \int_{-\infty}^{-\infty} f(\tau-t)g(t) dt
\end{align}

## Correlation
compute similarity while shifting signal g and keeping f stationary. For complex functions the complex conjugate is required. An correlation with itself (autocorrelation) produces a pure real signal.<br>
application: compute similarity

\begin{align}
(f \star g)(\tau) &= \int_{-\infty}^{-\infty} \overline{f}(t)g(t+\tau) dt
\end{align}

Gaußscher Integralsatz

\begin{align}
\int_V \vec{\nabla} \cdot \vec{F} \; d^{(n)}V = \oint_{S} \vec{F} \cdot \vec{n} \; d^{(n-1)}S
\end{align}

Stokscher Integralsatz

\begin{align}
\iint_{A} \vec{\nabla} \times \vec{F} \; d\vec{S} = \oint_{\partial A} \vec{F} \cdot \vec{n} \; dS
\end{align}

Greensche Formel 1

\begin{align}
\oint_{S} \Phi \frac{\partial \Psi}{\partial n} \; d^{(n-1)}S &= \oint_{S} ( \Phi \vec{\nabla} \Psi ) \cdot \vec{n} \; d^{(n-1)}S \\
&= \int_V \vec{\nabla} \cdot ( \Phi \vec{\nabla} \Psi ) \; d^{(n)}V \\
&= \int_V ( \Phi \vec{\nabla}^2 \Psi + \vec{\nabla} \cdot \Phi \vec{\nabla} \Psi ) \; d^{(n)}V
\end{align}

Greensche Formel 2

\begin{align}
\oint_{S} \Big( \Phi \frac{\partial \Psi}{\partial n} - \Psi \frac{\partial \Phi}{\partial n} \Big) \; d^{(n-1)}S = \oint_{S} \Phi \frac{\partial \Psi}{\partial n}  \; d^{(n-1)}S - \oint_{S} \Psi \frac{\partial \Phi}{\partial n} \; d^{(n-1)}S \\
= \int_V \big( \Phi \vec{\nabla}^2 \Psi + \vec{\nabla} \cdot \Phi \vec{\nabla} \Psi \big) \; d^{(n)}V - \int_V \big( \Psi \vec{\nabla}^2 \Phi + \vec{\nabla} \cdot \Psi \vec{\nabla} \Phi \big) \; d^{(n)}V \\
= \int_V \big( \Phi \vec{\nabla}^2 \Psi  - \Psi \vec{\nabla}^2 \Phi \big) \; d^{(n)}V
\end{align}

### Green funktion

A differential equation can be solved with help of Green functions

\begin{align}
&L U(x) = f(x) \qquad \text{(DGL)} \\
&L G(x,s) = \delta (x-s) \qquad \text{(def GF)} \\
&U(x) = \int G(x,s) f(s) ds \qquad \text{(solution)} \\
&L U(x) = \int L G(x,s) f(s) ds = \int \delta (x-s) f(s) ds = f(x)
\end{align}

Solution for the spatial helmholtz equation


\begin{align}
\underbrace{\big( \triangle + k^2 \big)}_{L} U &= \underbrace{0}_{f(x)}
\end{align}

is the Point Source at position zero:

<font color='red'>

\begin{align}
G(\vec{r}) = A \frac{e^{ik|\vec{r}|}}{|\vec{r}|}
\end{align}

</font>

proof:

\begin{align}
\triangle G(\vec{r}) &= \triangle A \frac{e^{ik|\vec{r}|}}{|\vec{r}|} \\
&= \frac{1}{r^2} \frac{\partial}{\partial r} \bigg( r^2 \frac{\partial}{\partial r} \bigg) A \frac{e^{ik|\vec{r}|}}{|\vec{r}|} \\
&= \frac{1}{r^2} \frac{\partial}{\partial r} \bigg( ikr^2 - r \bigg)  A \frac{e^{ik|\vec{r}|}}{|\vec{r}|} \\
&= - k^2  A \frac{e^{ik|\vec{r}|}}{|\vec{r}|} \\
&= - k^2 G(\vec{r})
\end{align}

so that

\begin{align}
\big( \triangle + k^2 \big) G(\vec{r}) &= \delta(\vec{r})
\end{align}
