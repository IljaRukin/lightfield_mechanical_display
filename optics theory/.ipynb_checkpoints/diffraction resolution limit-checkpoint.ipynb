## Abbe diffraction limited resolution

When an object is imaged the light passes trough a circular aperture with radius w. We use the fresnel integral to calculate the result.

\begin{align}
U(P) &= \frac{e^{-ikz}}{i z \lambda} e^{-\frac{ik}{2z} \big( x^2 + y^2 \big)} \int_{A,B} \Bigg[ \text{circ} \bigg( \frac{\sqrt{x^2 + y^2}}{w} \bigg) e^{-\frac{ik}{2z} \big( u^2 + v^2 \big)} \Bigg] e^{+\frac{ik}{2z} \big( xu + yv \big)} dudv \\
&= \frac{e^{ikz}}{i z \lambda} e^{\frac{ik r^2}{2z}} \Bigg[ 2 \frac{\text{J}(kwr/z)}{(kwr/z)} \Bigg]
\end{align}

The resulting Intensity is called the airy function

\begin{align}
I(P) = |U(P)|^2 = \bigg( \frac{2}{z \lambda} \frac{\text{J}(kwr/z)}{(kwr/z)} \bigg)^2
\end{align}

where $J(r)$ is the bessel function of first kind and order. This function smears out every point (and feature) of an image by convolution.
The first zero occurs at $r=\frac{0.61 \lambda z}{w}=\frac{1.22 \lambda z}{D}$.

<img src="abbe1.PNG" width='500px'>

If two points are observed, they can be observed seperated if one point lies inside the minimum of the second points airy function (Rayleigh-Criterion). The distance of the two points on the screen has to be $d>\frac{1.22 \lambda b}{D}$ (note z was renamed into b).



### microscope

<img src="abbe2.PNG" width='500px'>

We can use $\sin(\alpha) \approx \tan(\alpha) = \frac{D/2}{g}$ from wich follows with the Rayleigh-Criterion $\frac{d g}{b} >\frac{0.61 \lambda}{\sin(\alpha)}$. <br>
With the interception theorem (Strahlensatz) $\frac{d}{b}=\frac{\delta}{g}$, we get the resolution limit $\delta>\frac{0.61 \lambda}{\sin(\alpha)}$.<br>
In air the numerical aperture describes the maximal angle, at wich light can enter into the aperture of the lens. It is defined as $NA=\sin(\alpha)$. When using an optical dense medium the NA becomes $NA=\sin(\alpha) = \frac{n'}{n=1}\sin(\tilde{\alpha})$. For example if were to observe a sample with aperture angle $\alpha$ and place an optically dense material before the lens, then the light would be squished such that the actual angle $\tilde{\alpha}$ of the same light rays is smaller and the same aperture has a higher effective angle $\tilde{\alpha}$.<br>
For this reason the equation the resolution limit is modified into $\delta>\frac{0.61 \lambda}{n' \sin(\alpha)}$ for optically dense media before the lens.


### telescope

Rayleigh-Criterion $d>\frac{1.22 \lambda b}{D}$ with interception theorem (Strahlensatz) $\frac{d}{b}=\frac{\delta}{g}$ results in $\delta>\frac{1.22 \lambda g}{D}$. With $\theta \approx \tan(\theta) = \frac{\delta}{g}$ we get the angular resolution limit $\theta>\frac{1.22 \lambda}{D}$.


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as functions

r = np.linspace(-10,10,101)
airy = lambda r: np.divide( functions.jv(1, r) , r )
fig = plt.figure(figsize=(6,3),dpi=150)
plt.subplot()
plt.title('airy function',fontsize=18)
plt.scatter( r, airy(r), s=2 )
```
