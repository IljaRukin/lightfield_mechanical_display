## Kohärenz

### zeitlich Kohärenz

Interferenz einer phasenversetzer elektromagnetischer Welle $E_1(t)$ mit Intensität $I_0$ (z.B. Michelson-Morley-Interferometer)

\begin{align}
E(t)=E_1(t)+E_2(t)=E_1(t)+E_1(t+\tau)
\end{align}

\begin{align}
I &= <E^*E>_t \\
&= <(E_1+E_2)^*(E_1+E_2)>_t \\
&= <E_1^*E_1>_t + <E_2^*E_2>_t + <E_1^*E_2>_t + <E_2^*E_1>_t \\
&= 2I_0 + 2 Re \big\{ <E_1^*E_2>_t \big\} \\
&= 2I_0 + 2 Re \big\{ \Gamma(\tau) \big\}
\end{align}

mit der Selbstkohärenz
\begin{align}
\Gamma(\tau) &= < E_1(t) E_2(t) >_t \\
&= < E_1(t) E_1(t+\tau) >_t \\
&= \lim_{T \rightarrow \infty} \int_{-T/2}^{T/2} E_1(t) E_1(t+\tau) dt
\end{align}

bzw. dem Kohärenzgrad $|\gamma(\tau)| \leq 1$
\begin{align}
\gamma(\tau) &= \frac{\Gamma(\tau)}{\Gamma(0)} = \frac{\Gamma(\tau)}{I_0}
\end{align}

so dass
\begin{align}
I &= 2I_0 \big( 1 + Re \big\{ \gamma(\tau) \big\} \big)
\end{align}

wobei der Kontrast des Interferenzmusters gleich dem Kohärenzgrad ist
\begin{align}
V &= \frac{I_{max} - I_{min}}{I_{max} + I_{min}} \\
&= |\gamma(\tau)|
\end{align}

der Betrag des Kohärenzgrades ist somit bei zeitlich kohärententem licht 1 und bei zeitlich inkohärententem licht 0.


### örtliche Kohärenz (heuristisch)

<img src="doppelspalt.bmp" width="500px">

Interferenz einer örtlich ausgedehnten Lichtquelle am Beispiel des Youngschen Doppelspaltversuchs:

Punktlichtquelle bei S1: $E(r,t)=E_0 e^{ikr-iwt}$ ergibt ein deutliches Interferenzmuster
\begin{align}
I &= E^*E \\
&= I_0 \big( e^{-ik(r_1+l_1)} + e^{-ik(r_2+l_2)} \big)\big( e^{ik(r_1+l_1)} + e^{ik(r_2+l_2)} \big) \\
&= I_0 \big( 2 + e^{ik(-r_1-l_1+r_2+l_2)} + e^{-ik(-r_1-l_1+r_2+l_2)} \big) \\
&= I_0 \big( e^{\frac{ik}{2}(-r_1-l_1+r_2+l_2)} + e^{-\frac{ik}{2}(-r_1-l_1+r_2+l_2)} \big)^2 \\
&= 4 I_0 \cos^2\big(\frac{k}{2}(-r_1-l_1+r_2+l_2)\big) \\
&= 4 I_0 \cos^2\big(\frac{\pi}{\lambda}(-r_1-l_1+r_2+l_2)\big) \\
&= 4 I_0 \cos^2\big(\frac{\pi}{\lambda}\Delta l\big) \\
\end{align}
dabei sind $r_i$ die Weglängen von der Punktlichtquellen zu den Spaltern $i$ und $l_i$ die Weglängen von den Spalten zum jeweiligen Punkt auf dem Bildschirm.

Der Wegunterschied für eine Punktlichquelle $\Delta l $ kann angenähert werden:

\begin{align}
\Delta l &= r_2+l_2-r_1-l_1 \\
\frac{\Delta l}{d} &\approx \frac{l}{R} + \frac{y_0}{L}
\end{align}

Wenn beide Punktlichquellen S1,S2 vorhanden sind (ausgebreitete Lichtquelle), so überlagern sich beide Interferenzmuster und der Kontrast nimmt ab. In diesem Fall sind die Längen $r_1,r_2$ unterschiedlich lang. Das 0.te Maximum ($\Delta t = 0$) von S2 liegt bei $y_0=0$, denn $l=0$. Für S1 unterscheidet sich die Weglänge um $\Delta \big( \Delta l \big) = (r_1-r_2)$, so dass hier das Maximum veschoben ist:

\begin{align}
\Delta \big( \Delta l \big) &= d\Big( \frac{l}{R} + \frac{0}{L} \Big) - d\Big( \frac{l}{R} + \frac{\Delta y}{L} \Big) \\
&= d \frac{\Delta y}{L} \\
&= (r_1-r_2) \\
\Leftrightarrow \Delta y &= \frac{L}{d} (r_1-r_2)
\end{align}

Solanged der Wegunterschied sich maximal um $(r_1-r_2) < \frac{\lambda}{2}$ unterscheidet, was $\Delta y < \frac{\lambda L}{2d}$ entspricht, kann ein Interferenzmuster beobachtet werden. Durch Punkte zwischen S1 und S2 wird diese Bedingung etwas aufgeweicht.<br>

Ein wichtige Erkenntnis ist, dass die örtliche Kohärenz nicht nur von der Ausdehnung der Lichtquelle, sondern auch von dem Spaltabstand $d$ und dem Schirmabstand $L$ abhängt.

### örtliche Kohärenz

Die räumliche Kohärenzfunktion aus Überlagerung der Amplituden zweier 
Punktquellen $E_1(x,t),E_2(x,t)$ (Spalte des Dopperspaltexperiments)
auf einem Schirm mit Koordinate $\alpha$:

\begin{align}
I &= <E(\alpha)^*E(\alpha)> \\
&= <(E_1(r_1)+E_2(r_2))^*(E_1(r_1)+E_2(r_2))> \\
&= <E_1^*E_1> + <E_2^*E_2> + <E_1^*E_2> + <E_2^*E_1> \\
&= I_1 + I_2 + 2 Re \big\{ <E_1^*E_2> \big\} \\
&= I_1 + I_2 + 2 Re \big\{ \Gamma(r_1,r_2) \big\} \\
&= I_1 + I_2 + 2 Re \big\{ <A_1 e^{-ik_1r_1} \cdot A_2 e^{ik_2r_2}> \big\} \\
&= I_1 + I_2 + 2 \sqrt{I_1 I_2} Re \big\{ <e^{i(k_2r_2-k_1r_1)}> \big\} \\
&= \big( I_1 + I_2 \big) \bigg( 1 + 2 \underbrace{\frac{2 \sqrt{I_1 I_2}}{I_1 + I_2}}_{\beta} \underbrace{Re \big\{ <e^{i(k_2r_2-k_1r_1)}> \big\}}_{\gamma(r_1,r_2) = \frac{\Gamma(r_1,r_2)}{2\sqrt{I_1 I_2}}} \bigg) \\
&= I_1 + I_2 + 2 \sqrt{I_1 I_2} Re \big\{ <e^{i(k_2r_2-k_1r_1)}> \big\} \\
\end{align}

wobei der Kontrast des Interferenzmusters gleich dem Kohärenzgrad mit Konstante $\beta$ ist

\begin{align}
V &= \frac{I_{max} - I_{min}}{I_{max} + I_{min}} \\
&= |\gamma(\tau)| \beta
\end{align}

und für $I_1 = I_2$:

\begin{align}
V &= |\gamma(\tau)|
\end{align}

der Betrag des Kohärenzgrades ist somit bei örtlich kohärententem licht 1 und bei örtlich inkohärententem licht 0.

Dieses mal ist $<>$ ein Mittel über ein Enseamble. Für zwei Ebene Wellen, wie oben beschrieben, ist der Kohärenzgrad immer 1. Im Falle, dass die Amplituden $E_1(r,t),E_2(r,t)$ aus zwei Spalten bei $x_1,x_2$ stammt und Ursprünglich aus einer örtlich ausgedehnten Quelle mit Koordinate $\xi$ kam, kann der Kohärenzgrad kleiner ausfallen. Große Abstände zwischen den Ebenen $\xi,x,\alpha$ werden mit einem 4f Aufbau simuliert.

<img src='citter_zernike.png' width='500px'>

Dann wird die Intensität $E_i$ mit einem Integral oder einer Summe über die Lichtquelle berechnet:

\begin{align}
E_i = \sum_\xi U(\xi) e^{ik \frac{x_i \xi}{f}}
\end{align}

Die Korrelationsfunktion ist in diesem Fall

\begin{align}
\Gamma(r_1,r_2) &= <E_1(r_1)^*E_2(r_2)> \\
&= < \sum_\xi U(\xi) e^{-ik \frac{x_1 \xi}{f}} \cdot \sum_{\xi'} U(\xi') e^{ik \frac{x_2 \xi'}{f}} \cdot e^{i(k_2r_2-k_1r_1)} > \\
&= 2 < \sum_{\xi=\xi'} I(\xi) e^{-ik \frac{(x_2-x_1) \xi}{f}} \cdot e^{i(k_2r_2-k_1r_1)} > + 2 < \sum_{\xi \neq \xi'} U(\xi)U(\xi') e^{-ik \frac{x_1 \xi}{f}} U(\xi') e^{ik \frac{x_2 \xi'}{f}} \cdot e^{i(k_2r_2-k_1r_1)} >
\end{align}

Die zeitliche Korrelation zwischen Licht aus zwei Unterschiedlichen Punkten $\xi \neq \xi'$ kann dann zu Null angenähert werden, wenn das Licht aus einer zeitlich inkohärenten Quelle stammt. Zusätzlich wird die Summe zum Integral:

\begin{align}
\Gamma(r_1,r_2) &= 2 < \sum_{\xi=\xi'} I(\xi) e^{-ik \frac{(x_2-x_1) \xi}{f}} \cdot e^{i(k_2r_2-k_1r_1)} > \\
&= <\int I(\xi) e^{ik \frac{(x_2-x_1) \xi}{f}} d\xi \cdot e^{i(k_2r_2-k_1r_1)} > \\
&= I_0 2a \cdot \text{sinc} \big( k \frac{(x_2-x_1)}{f} \big) \cdot e^{i(k_2r_2-k_1r_1)} >
\end{align}

Das Ergebnis ist, das die Lichtquelle fouriertransformiert auf den Spalten ankommt und von dort aus mit $e^{i(k_2r_2-k_1r_1)}$ zum Schirm propagiert wird.

### raum-zeitliche Kohärenzfunktion

Definition

\begin{align}
\Gamma_{1,2}(\tau) &= \Gamma(r_1,r_2,\tau) \\
&= <E^*(r_1,t)E(r_2,t+\tau)>_t 
\end{align}

Daraus lässt sich der Wechselseitige Kohärenzgrad definieren:

\begin{align}
\gamma_{1,2}(\tau) &= \frac{\Gamma_{1,2}(\tau)}{\sqrt{\Gamma_{1,1}(\tau=0)\Gamma_{2,2}(\tau=0)}}
\end{align}

## Spektrale Leistungsdichte

Aus der Selbstkohärenz
\begin{align}
\Gamma(\tau) &= < E(t) E^*(t+\tau) >_t \\
&= \lim_{T \rightarrow \infty} \int_{-T/2}^{T/2} E(t) E^*(t+\tau) dt
\end{align}

lässt sich die Spektrale Leistungsdichte $S(\omega)=|E(\omega)|$ berechnen:
\begin{align}
S(\omega) = \int_{\infty}^{\infty} \Gamma(\tau) e^{-i \omega \tau} df
\end{align}

Beweis (siehe Whiteboard: Wiener-Chinschin-Theorem), indem die zeitliche Kreuzkorrelation zur Multiplikation im Zeit-Frequenzraum wird.
