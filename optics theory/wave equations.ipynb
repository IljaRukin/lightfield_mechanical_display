## electric & magnetic field in matter

\begin{align}
\vec{D} &= \epsilon_0 \vec{E} + \vec{P} = \epsilon_0 \vec{E} + \epsilon_0 \chi_e \vec{E} = \epsilon_0 \epsilon_r \vec{E} \\
\vec{B} &= \mu_0 \vec{H} + \mu_0 \vec{M} = \mu_0 \vec{H} + \mu_0 \chi_m \vec{H} = \mu_0 \mu_r \vec{H}
\end{align}

with
\begin{align}
\epsilon_r = 1 + \chi_e \\
\mu_r = 1 + \chi_m
\end{align}

and the refractive index
\begin{align}
n = \sqrt{\epsilon_r \mu_r} = \frac{c}{v} \\
c = \frac{1}{\epsilon_0 \mu_0} \\
\end{align}


## Maxwell-Equations

\begin{align}
\oint_A \vec{D} d \vec{S} = \iiint_V \rho dV \qquad &\Leftrightarrow \qquad \vec{\nabla} \cdot \vec{D} = \rho \qquad \text{Coulombsches Induktionsgesetz (Gauss's law)} \\
\oint_A \vec{B} d \vec{S} = 0 \qquad &\Leftrightarrow \qquad \vec{\nabla} \cdot \vec{B} = 0 \qquad \text{keine magnetischer Monopole (Gauss's law for magnetism)} \\
\oint_C \vec{E} d \vec{l} = - \frac{d}{dt} \iint_A \vec{B} d\vec{S} \qquad &\Leftrightarrow \qquad \vec{\nabla} \times \vec{E} = - \frac{\partial}{\partial t} \vec{B} \qquad \text{Faradaysches Induktionsgesetz (Faraday equation)} \\
\oint_C \vec{H} d \vec{l} = \iint_A \vec{J} d \vec{S} + \frac{d}{dt} \iint_A \vec{D} d \vec{S} \qquad &\Leftrightarrow \qquad \vec{\nabla} \times \vec{H} = \vec{J} + \frac{\partial}{\partial t} \vec{D} \qquad \text{Amperesches Verkettungsgesetz (Amp√®re's circuital law)} \\
\end{align}

the charge conservation equation

\begin{align}
\vec{\nabla} \cdot \vec{J} = - \frac{\partial}{\partial t} \rho
\end{align}

and ohms law

\begin{align}
\vec{J} = \sigma \vec{E}
\end{align}

### Wave equation

To derive the maxwell equations we apply the curl operator on the Faraday equation (3) and insert Amepres circuital law (4):

\begin{align}
\vec{\nabla} \times \vec{\nabla} \times \vec{E} &= - \vec{\nabla} \times \frac{\partial}{\partial t} \vec{B} \\
&= - \frac{\partial}{\partial t} \vec{\nabla} \times \vec{B} \\
&= - \mu_0 \mu_r \frac{\partial}{\partial t} \vec{\nabla} \times \vec{H} \\
&= - \mu_0 \mu_r \frac{\partial}{\partial t} \Big( \vec{J} + \frac{\partial}{\partial t} \vec{D} \Big) \\
&= - \mu_0 \mu_r \frac{\partial}{\partial t} \Big( \sigma \vec{E} + \epsilon_0 \epsilon_r \frac{\partial}{\partial t} \vec{E} \Big) \\
&= - \sigma \mu_0 \mu_r \frac{\partial}{\partial t} \vec{E} - \epsilon_0 \epsilon_r \mu_0 \mu_r \frac{\partial^2}{\partial t^2} \vec{E} \\
\end{align}

with $\vec{\nabla} \times \big( \vec{\nabla} \times \vec{E} \big) = \vec{\nabla} \cdot \big( \vec{\nabla} \cdot \vec{E} \big) - \triangle \vec{E}$ and the electric field beeing a result of a scalar field $\vec{E} = -\vec{\nabla} \Phi$ such that $\vec{\nabla} \times \vec{E} = - \vec{\nabla} \times \vec{\nabla} \Phi = 0$ we get:

\begin{align}
\triangle \vec{E} &= \sigma \mu_0 \mu_r \frac{\partial}{\partial t} \vec{E} + \epsilon_0 \epsilon_r \mu_0 \mu_r \frac{\partial^2}{\partial t^2} \vec{E} \\
\end{align}

this equation is valid for isotropic, charge-free and non-dispersive media. For inhomogeneous media we get an additional term $-\vec{\nabla} \cdot \big( \frac{\vec{\nabla} \epsilon_r}{|\epsilon_r|} \cdot \vec{E} \big)$ on the right hand side.

for non-conductive, non-magnetic, homogeneous media $\sigma=0,\mu_r=1,\vec{\nabla},\epsilon_r=0$ this equation simplifies considerably:

\begin{align}
\frac{1}{c^2} \frac{\partial^2}{\partial t^2} E(\vec{r},t) = \triangle E(\vec{r},t)
\end{align}



### Helmholtz equation

Applying separation

\begin{align}
E(\vec{r},t) = U(\vec{r}) \cdot T(t) 
\end{align}

we get

\begin{align}
\frac{1}{c^2 U} \frac{\partial^2}{\partial t^2} U &= \frac{1}{T} \triangle T \\
&\stackrel{!}{=} - k^2
\end{align}

The factor $- k^2$ is in agreement with a fourier transform of the laplace operator. This makes the result for $E(\vec{r},t) = U(\vec{r}) \cdot T(t) $ equally valid in the space and spatial-frequency domain.

rearranging with use of $\omega=kc$ get's us

\begin{align}
\frac{\partial^2}{\partial t^2} U &= - \omega^2 U \qquad \text{temporal}\\
\triangle T &= - k^2 T \qquad \text{spatial}
\end{align}


### plane waves

solution in kartesian coordinates.
The results are <b>plane waves</b>:

\begin{align}
U(\vec{r},t) &= U_0 e^{i(\vec{k}_0 \vec{r} - \omega t)} \\
U(\vec{k}_0,t) &= U_0 \delta(\vec{k}-\vec{k}_0) e^{- (i \omega t)}
\end{align}

with one constant frequency/direction $k_0 = |\vec{k}_0|$ and distance vector $\vec{r}=(x,y,z)$.

For $\hat{e}_z$ parallel to $\vec{k}$ follows for $U(\vec{r},t)$ along the z-axis:

\begin{align}
U(\vec{r},t) = U(z,t) = U_0 e^{i(k_z z - \omega t)}
\end{align}



### spherical waves

solution in spherical coordinates with only radial component $r \hat{e}_r$.
The results are <b>spherical waves</b>:

\begin{align}
U(r,t) &= U_0 \frac{e^{i(k'r - \omega t)}}{r} \\
&= U_0 \frac{e^{i(k'\sqrt{x^2+y^2+z^2} - \omega t)}}{\sqrt{x^2+y^2+z^2}} \\
\end{align}

with constant frequency/direction $k'=constant$ parallel to the distance vector $r=\sqrt{x^2+y^2+z^2}$, such that $\vec{r} \cdot \vec{k}' = r \cdot k'$

It can be decomposed into plane waves by fourier transform:

(by rotating the coordinate system, such that $\vec{k} \vec{r} = k r \cos(\theta)$ and substituion of $o = - \cos(\theta)$ and $do = \sin(\theta) d\theta)$)

\begin{align}
U(k,t) &= \iiint_{-\infty}^{\infty} U_0 \frac{e^{i(k'\sqrt{x^2+y^2+z^2} - \omega t)}}{\sqrt{x^2+y^2+z^2}} e^{i(k_x x + k_y y + k_z z)} dx dy dz \\
&= \int_{0}^{\infty}\int_{0}^{2 \pi}\int_{0}^{\pi} U_0 \frac{e^{i(k'r - \omega t)}}{r} e^{i \vec{k} \vec{r}} r^2 \sin(\theta) d\theta d\varphi dr \\
&= \int_{0}^{\infty}\int_{0}^{2 \pi}\int_{0}^{\pi} U_0 e^{i(k'r - \omega t)} e^{i k r \cos(\theta)} r \sin(\theta) d\theta d\varphi dr \\
&= \int_{0}^{\infty}\int_{0}^{2 \pi}\int_{-1}^{1} U_0 e^{i(k'r - \omega t)} e^{-i k r o} r do d\varphi dr \\
&= 2 \pi \int_{0}^{\infty} U_0 e^{i(k'r - \omega t)} \frac{e^{-i k r o}}{-ikr} \Bigg|_{-1}^{1} r dr \\
&= 2 \pi U_0 \int_{0}^{\infty} \frac{e^{i(k'r - \omega t)}}{ik} \Big( e^{i k r} - e^{-i k r} \Big) dr \\
&= \frac{2 \pi U_0}{ik} e^{-i \omega t} \int_{0}^{\infty} \Big( e^{i (k' + k) r} - e^{i (k' - k) r} \Big) dr \\
&= \frac{2 \pi U_0}{ik} e^{-i \omega t} \delta(k - k') dr \\
\end{align}

where the frequencies/directions $\vec{k}$ in 3D fourier space are located on the surface of a sphere with radius $k = |\vec{k}'| = k'$.

for particular direction $\vec{e}_z$ along the z-axis the wave in a plane xy is:

\begin{align}
U(x,y,z,t) &= U_0 \frac{e^{i\big(k'\sqrt{x^2+y^2+z^2} - \omega t\big)}}{r} \\
&= U_0 \frac{e^{i\big(k'z\sqrt{1+\big(\frac{x}{z}\big)^2+\big(\frac{y}{z}\big)^2} - \omega t\big)}}{r} \\
&= U_0 \frac{e^{i\big(z\sqrt{k'^2+\big(k'\frac{x}{z}\big)^2+\big(k'\frac{y}{z}\big)^2} - \omega t\big)}}{r} \\
\end{align}



### spherical waves (paraxial approximation)

We approximate a spherical wave for small $x,y$ with a Taylor series of the square root and dropping the phase factor $ik'z$ at the end:

\begin{align}
U(r,t) &= U_0 \frac{e^{ik'r}}{r} \\
&\approx U_0 e^{ik'r} \\
&= U_0 e^{ik'\sqrt{x^2+y^2+z^2}} \\
&= U_0 e^{ik' z \sqrt{1 + \frac{x^2 + y^2}{z^2}}} \\
&\approx U_0 e^{ik' z \Big(z + \frac{x^2 + y^2}{2z^2}\Big)} \\
&\approx U_0 e^{ik' \frac{x^2 + y^2}{2z} } \\
\end{align}


### spherical waves (free space impulse response)

This can also be done the other way around.
It's possible to use the solution to the helmholtz equation in cartesian coordinates, say the plane waves and integrate over all directions $k$. We pick a monochromatic $k=\sqrt{k_x^2+k_y^2+k_z^2}$. Since we are interested in the propagation of the wave in z-direction we rewrite the plane waves

\begin{align}
e^{i\vec{k}\vec{r}} &= e^{i(k_x x+k_y y+k_z z)} \\
&= e^{i(k_x x+k_y y+k_z z)} \\
&= e^{i\big(k_x x+k_y y+z\sqrt{k^2-k_x^2-k_y^2}\big)} \\
\end{align}

and integrate over $k_x,k_y$:

\begin{align}
U(\vec{r}) &= \iint U(k_x,k_y) e^{i\big(k_x x+k_y y+\sqrt{k^2-k_x^2-k_y^2} z\big)} dk_x dk_y \\
&= \mathcal{F} \bigg( U(k_x,k_y) e^{iz\sqrt{k^2-k_x^2-k_y^2}} \bigg)_{k_x,k_y} \\
&= \mathcal{F} \bigg( U(k_x,k_y) e^{ikz\big(1-\frac{k_x^2-k_y^2}{2k^2}\big)} \bigg)_{k_x,k_y} \\
\end{align}

from that we get the impulse response for a EM-wave $U(k_x,k_y)$ in free space by $H(k_x,k_y) = e^{iz\sqrt{k^2-k_x^2-k_y^2}} \approx e^{ikz\big(1-\frac{k_x^2-k_y^2}{2k^2}\big)}$ in fourier space or $h(x,y) = \mathcal{F} \big( H(k_x,k_y) \big)_{k_x,k_y} = \frac{ e^{i k \big( z -\frac{x^2 + y^2}{2z} \big)} }{ \sqrt{\frac{z}{ik}}}$ in real space.

\begin{align}
U(x,y,z) = U(x,y,0) \circledast \frac{ e^{i k \big( z -\frac{x^2 + y^2}{2z} \big)} }{ \sqrt{\frac{z}{ik}} }
\end{align}

It's often said, that frequencies $k^2 < k_x^2 + k_y^2$ are exponentially decaying (evanescent waves), since the root of the impulse response becomes complex and the response itself a decaying exponential function. In fact a monochromatic wave with wave vector $k$ or it's linear combinations can not in principle represent an image with finer resolution (higer k) than it's own wavelength $\lambda = \frac{2\pi}{k}$ permits, since the image it'self is composed of a sum of this particular waves.



### peak in real space
if we write a plane wave in fourier space (let's call it a frequency plane wave). It has only one position $x_0$ (delta peak) in real space instead of one direction/frequency $k_0$: 

\begin{align}
U(k,t) = e^{ikx_0} \\
U(x,t) = \delta(x-x_0)
\end{align}
